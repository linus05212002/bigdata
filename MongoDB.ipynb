```python
#"$or":
    
cursor = db.courses.find({"$or": [{'lecturer.name': 'Frank Habermann'},{'lecturer.name': "Markus Löcher"}]})

for document in cursor:

    result = document
    pprint(result)
    print('\n',result['title'],result['semester'],'\n','-'*30)
   
```

```python
cursor = db.courses.find({'semester':{'$gt':1}})

for document in cursor:

    result = document
    pprint(result)
    print('\n',result['title'],result['semester'],'\n','-'*30)
```

```python
laureates =  requests.get('http://api.nobelprize.org/v1/laureate.json').content
```

```python
prizes =  requests.get('http://api.nobelprize.org/v1/prize.json').content
```

```python
# Create local "nobel" database on the fly 
db = client["nobel"]
db.prizes.drop()
db.laureates.drop()
# API documented at https://nobelprize.readme.io/docs/prize 
for collection_name in ["prizes", "laureates"]:
    singular = collection_name[:-1] # the API uses singular
    response = requests.get( "http://api.nobelprize.org/v1/{}.json".format(singular)) 
    documents = response.json()[collection_name] 
    # Create collections on the fly 
    db[collection_name].insert_many(documents)
```

```python
result = db.laureates.find_one()
print(result['firstname'])
```

```python
db.laureates.count_documents({'gender':'female'})
```

```python
#distinct: distinct list only all distinct entries.

db.laureates.distinct("bornCountry", {"bornCountry": {"$regex": "Germany"}})
```

```python
#How many laureates are from Germany? 

db.laureates.count_documents({"bornCountry": {"$regex": "Germany"}})
```

```python
answer = db.laureates.find({"bornCountry": {"$regex": "Germany"},'prizes.category': 'physics'})

for doc in answer:
    pprint(doc["prizes"][-1])
    print(doc['firstname'],",",doc['surname'],'\n','-'*30)
```

```python
answer = db.laureates.find({'gender': 'female'}).sort([("prizes.year",1)])

for doc in answer:
    pprint(doc["prizes"][-1])
    print(doc['firstname'],",",doc['surname'],'\n','-'*30)
```
